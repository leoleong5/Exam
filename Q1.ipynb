## Introduction

A text file can be subjected to various analyses using this Python script. The path to the text file the user wants to analyse must be provided.



How to Use:


When you first RUN the application, it will ask you for the full path to the text file you wish to evaluate. A selection of analysis possibilities is presented by the programme after you have entered the path. By entering the corresponding number and clicking enter, you can choose the desired choice.


Analysis Options:

Count the number of lines: The text file's line count is returned by this parameter.

Count the total number of words: The total amount of words in the text file is returned by this option. It considers any string of letters divided by a space or a line break to be a word.

Identify the frequency of each word in the text: This option gives the frequency of each word in the text file.

Identify the top 5 most common words in the text: The top five terms in the text file are returned together with their frequency when this option is selected.

Exit: This option ends the program.

The programme will shut down and display an error message if the wrong file path is entered. The programme will display an error notice and ask the user to enter a valid option number again if an invalid option number is entered.

## Program

````python
## Code from ChatGPT4
import collections
import re

def analyze_file():
    filepath = input("Please enter the path to the text file: ")

    try:
        with open(filepath, 'r') as file:
            content = file.read()

    except FileNotFoundError:
        print(f"No such file or directory: '{filepath}'")
        return

    while True:
        print("\nChoose an analysis option:")
        print("1. Count the number of lines")
        print("2. Count the total number of words")
        print("3. Count the total number of characters (including and excluding whitespace)")
        print("4. Identify the frequency of each word in the text")
        print("5. Identify the top 5 most common words in the text")
        print("6. Exit")

        option = input("Enter your choice: ")

        if option == '1':
            print(f"Number of lines: {len(content.splitlines())}")
        elif option == '2':
            print(f"Total number of words: {len(content.split())}")
        elif option == '3':
            print(f"Total number of characters (including whitespace): {len(content)}")
            print(f"Total number of characters (excluding whitespace): {len(content.replace(' ', ''))}")
        elif option == '4':
            words = re.findall(r'\b\w+\b', content.lower())
            word_count = collections.Counter(words)
            print("Word frequency:")
            for word, count in word_count.items():
                print(f"{word}: {count}")
        elif option == '5':
            words = re.findall(r'\b\w+\b', content.lower())
            word_count = collections.Counter(words)
            common_words = word_count.most_common(5)
            print("Top 5 most common words:")
            for word, count in common_words:
                print(f"{word}: {count}")
        elif option == '6':
            break
        else:
            print("Invalid option. Please try again.")

if __name__ == "__main__":
    analyze_file()
````
